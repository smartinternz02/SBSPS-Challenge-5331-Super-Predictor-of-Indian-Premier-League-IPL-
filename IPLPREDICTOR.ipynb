# Preprocessing

```python
#label encoding for venues, batting and bowling teams
#dataset=pd.get_dummies(dataset,drop_first=True)
from sklearn.preprocessing import  LabelEncoder
le= LabelEncoder()
dataset["venues"]=le.fit_transform(dataset["venue"])
dataset["batting_team"]=le.fit_transform(dataset["bat_team"])
dataset["bowling_team"]=le.fit_transform(dataset["bowl_team"])
```

```python
dataset.drop(['bat_team','bowl_team','venue'], axis=1, inplace=True)
```

```python
#rearranging columns of dataset
updated_dataset=dataset[['date', 'venues','batting_team', 'bowling_team', 'runs', 'wickets', 'overs', 'runs_last_5', 'wickets_last_5', 'total']]
```

```python
#converting date in string to datetime object so that data can be splitted according to values of year
from datetime import datetime
updated_dataset['date'] = dataset['date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
```

```python
#train-test splitting
train=updated_dataset[updated_dataset['date'].dt.year<2015]
test=updated_dataset[updated_dataset['date'].dt.year>=2015]
```

```python
train.shape
```

```python
test.shape
```

```python
X=updated_dataset.drop(labels='total',axis=1)
X.drop(labels='date',axis=1,inplace=True)
Y=updated_dataset['total'].values
```

```python
X.head()
```

```python
X_train=train.drop(labels='total',axis=1)
X_test=test.drop(labels='total',axis=1)
Y_train=train['total'].values
Y_test=test['total'].values
```

```python
#dropping date column from training set
X_train.drop(labels='date',axis=1,inplace=True)
X_test.drop(labels='date',axis=1,inplace=True)
```

```python
#model1
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
model1=LinearRegression(fit_intercept=True,normalize=False,copy_X=True,n_jobs=None,positive=False)
model1.fit(X_train,Y_train)
model1_pred = model1.predict(X_test)
model1_predt=model1.predict(X_train)
model1_rmse_score = np.sqrt(mean_squared_error(Y_test, model1_pred))
model1_rmse_scoret=np.sqrt(mean_squared_error(Y_train,model1_predt))
model1_rsquared_score = r2_score(Y_test, model1_pred)
model1_mae=mean_absolute_error(Y_test,model1_pred)
print("MODEL 1 LINEAR REGRESSION")
print('RMSE score for test:', model1_rmse_score)
print('RMSE score for train:', model1_rmse_scoret)
print('R2 score:', model1_rsquared_score)
print('MAE score:', model1_mae)
print('Variance score test:', model1.score(X_test, Y_test))
print('Variance score train:', model1.score(X_train, Y_train))
sns.regplot(Y_test, model1_pred);
```

```python
#crossvalidationofmodel1
from sklearn.model_selection import cross_val_score
from sklearn import model_selection
model1_scores= model_selection.cross_val_score(model1, X, Y, scoring="neg_mean_squared_error", cv=model_selection.KFold(n_splits=10, random_state=6,shuffle=True))
model1_rmse_scores=np.sqrt(-model1_scores)
print("MODEL 1 LINEAR REGRESSION CROSS VALIDATION")
print("RMSE Scores:",model1_rmse_scores)
print("RMSE Mean:",model1_rmse_scores.mean())
print("RMSE Standard Deviation:",model1_rmse_scores.std())
model1_r2_scores= model_selection.cross_val_score(model1, X, Y, scoring="r2", cv=model_selection.KFold(n_splits=10, random_state=5,shuffle=True))
print("R2 SCORES MEAN", model1_r2_scores.mean())
```
